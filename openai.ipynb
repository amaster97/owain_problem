```python
N = 100
p = .5
e = .0
S = '(Statement is True) AND (Statement is all lowercase letters) AND (Statement discusses finance).'


response = client.chat.completions.create(
  model="gpt-4-1106-preview",
  messages=[
    {"role": "system", "content": "I will input parameters in this format: {N, p, e, RULE}.  You will generate N statements.  With probability p they will follow RULE and with probability 1-p they will not follow RULE.  Your output should be {Input: STATEMENT | Label: True/False}, where STATEMENT is what is generated; and True if the rule is followed and False if it isn't.  However with probability e, the True|False will be mislabelled."},
    {"role": "user", "content": "{}, {}, {}, {}".format(N, p, e, S)},
    ]
)

statements = response.choices[0].message.content

print(statements.split('\n')[0:10])
```
